In Seaborn, the hue parameter is used to add a third dimension to a plot by grouping data points based on the values of a categorical variable. When you specify a variable for hue, Seaborn assigns different colors to the data points corresponding to different levels of that variable. This is particularly useful for visualizing how data points are distributed or how they relate to other variables within different subgroups of your data.

```python
sns.relplot(x = 'total_bill' , y ='tip'  , data = df1 ,style = 'time',hue = 'smoker')
#style is also same as hue but uses symbols instead of colours 
#style='time': This tells Seaborn to use different marker styles (e.g., circles, squares, triangles) to represent the time variable, which has categories 'Lunch' and 'Dinner'.
```

In the below code: In Seaborn, the col parameter is used to create multiple plots, each representing a subset of the data based on the values of a categorical variable. Each subplot (column) represents a distinct category of the variable specified by col, allowing for easy comparison between different categories.

```python
sns.relplot(x = 'total_bill' , y ='tip' , col= 'time',data =df1)
#col='time': This tells Seaborn to create separate plots (subplots) for each unique category of the time variable. In the plot:
#One subplot will represent data for lunch.
#Another subplot will represent data for dinner.
#By using col, you can easily compare the relationship between total_bill and tip separately for lunch and dinner times.
```

In Seaborn, the sns.catplot function is used to create categorical plots, which are useful for visualizing the distribution of a categorical variable. By default, sns.catplot creates a strip plot, but you can customize it to create other types of categorical plots like box plots, violin plots, bar plots, etc., by using the kind parameter.

```python
sns.catplot(x = 'day' , y = 'total_bill', palette='Set2', data = df1)
#palette='Set2': Specifies a predefined color palette. Seaborn provides several built-in palettes such as 'Set1', 'Set2', 'Set3', 'pastel', 'dark', 'colorblind', etc.
```

```python
sns.catplot(x = 'day' , y = 'tip', data = df1)
```

For below code:


In Seaborn, the kind='swarm' parameter in sns.catplot creates a swarm plot. A swarm plot is a type of scatter plot that arranges the data points to avoid overlap, which helps to show the distribution of data points for different categories. Unlike strip plots, swarm plots ensure that all data points are visible and not obscured by others.

A swarm plot is a categorical scatter plot where the data points are adjusted along the categorical axis (x-axis in this case) to avoid overlapping. This gives a clear view of the distribution of the data points.

```python
sns.catplot(x = 'day' , y = 'tip', data = df1 , kind = 'swarm')
```

```python
sns.catplot(x = 'day' , y = 'tip',kind = 'box' , data = df1) #boxplot
```

```python
sns.catplot(x = 'day' , y = 'tip',kind = 'box' , hue = 'sex', data = df1)
```

Creating a pair plot using Seaborn's sns.pairplot function is an excellent way to visualize pairwise relationships in a dataset. This is particularly useful for identifying correlations, patterns, and potential outliers in multivariate data. Let's use the iris dataset to demonstrate this.

Pair Plot Components:
A pair plot shows scatter plots for each pair of variables in the dataset, along with histograms or density plots along the diagonal, which represent the distribution of each variable.

Scatter Plots: Each off-diagonal plot is a scatter plot showing the relationship between two variables.

Histograms: Diagonal plots are histograms (or density plots) showing the distribution of individual variables.
