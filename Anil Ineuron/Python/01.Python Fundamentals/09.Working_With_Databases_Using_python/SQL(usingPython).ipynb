# Group By Operation

```python
cursor.execute("select count(col1) from testsudhineuron.glassdata group by col1")
cursor.fetchall()
```

```python
cursor.execute("select col1, count(col1)  from testsudhineuron.glassdata group by col1")
#yha pr hm count kr rhe hai ki col1 mei particular value kitne baar occur hua hai
#jaise 1 -> 7 times aya hai, 2 -> 6 times etc
```

```python
cursor.fetchall()
```

```python
cursor.execute("select col1, col11  from testsudhineuron.glassdata order by col1 desc")
cursor.fetchall()

#bydefault order by gives output in ascending (asc) order
```

```python
cursor.execute("select col1, count(*) from testsudhineuron.glassdata group by col1 order by col1 desc")
cursor.fetchall()

#count(*) likhu chahe count(col1), working of both is same
```

```python
#or command
cursor.execute("select col1 , col2 , col11 from testsudhineuron.glassdata where col1 = 11 or  col11 = 2")
cursor.fetchall()
```

```python
#and command
cursor.execute("select col1 , col2 , col11 from testsudhineuron.glassdata where col1 = 11 and  col11 = 2")
cursor.fetchall()
```

```python
#like command
cursor.execute("select * from testsudhineuron.glassdata where  col2 like '%.51%'")
cursor.fetchall()
```

```python
cursor.execute("select * from testsudhineuron.glassdata  order by col2")
cursor.fetchall()
```

```python
#fetching record from particular query
#nested query
#giving alias "sudh" is important
cursor.execute("select * from (select * from testsudhineuron.glassdata  where col1 > 112) as sudh  where col11 = 3")
cursor.fetchall()
```

```python
cursor.execute("select * from  testsudhineuron.glassdata order by col1")
cursor.fetchall()
```

```python
#IN & HAVING command
#query for showing all the records where occurence of particular value in col1 is less than 6
cursor.execute("SELECT * FROM testsudhineuron.glassdata WHERE col1 IN (SELECT col1 FROM testsudhineuron.glassdata GROUP BY col1 HAVING COUNT(col1) < 6) order by col1")
cursor.fetchall()
```
