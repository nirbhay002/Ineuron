### Must see this too:
In statistics, the method for calculating percentiles can vary, and the approach you use depends on the context or the specific statistical software being employed. There are two common approaches to percentile calculation:

### 1. **Linear Interpolation (Default Method):**

This is the method commonly used in NumPy's `np.percentile`, and its the one we used earlier. Heres the reasoning:

- **Formula:**

  

  - \(P\) is the percentile you want to compute.
  - \(N\) is the number of data points.

  The term \((N - 1)\) is used because we are computing the position relative to a zero-based index, which is typical in programming and data manipulation. 

- **Why not \(N + 1\)?** Using \(N - 1\) provides a position in the list that accommodates zero-based indexing, making the computation straightforward and consistent with the way lists are indexed in programming.

### 2. **Rank-Based Method (Traditional Approach):**

This approach is often used in mathematical statistics and might be the method youve encountered. The formula for this method is:

- **Formula:**

  

  - \(P\) is the percentile.
  - \(N\) is the number of data points.

  This method calculates a rank thats 1-based (starting from 1), which can make the interpretation more intuitive for some statistical contexts.

  **Note:** When using this method, if the rank is not an integer, you often interpolate between the values at the ranks surrounding the calculated position.

### Summary of Both Methods:

- **NumPys Method (Linear Interpolation):**

  

  - Often used in programming.
  - Handles zero-based indexing.

- **Rank-Based Method:**

  

  - Often used in traditional mathematical statistics.
  - Handles 1-based ranking.

Each method has its merits and is used depending on the specific context or software convention. When in doubt, its important to check the documentation or the method used by the software you're working with.

````python
#python code for variance for sample size
def variance(data):
    n=len(data)
    ## mean of the data
    mean=sum(data)/n
    ## variance
    deviation=[(x - mean) ** 2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
````
