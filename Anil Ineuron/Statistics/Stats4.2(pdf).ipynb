```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import seaborn as sns
import pylab

def plot_data(sample):
    """
    Plots a histogram with a KDE and a Q-Q plot to assess if the sample data follows a normal distribution.

    Parameters:
    sample (array-like): A sequence of numerical data points to be analyzed.

    The function creates a figure with two subplots:
    1. The first subplot displays a histogram of the sample data with a Kernel Density Estimate (KDE) overlaid.
    2. The second subplot displays a Q-Q plot comparing the quantiles of the sample data to the quantiles of a normal distribution.
    
    The Q-Q plot helps to determine if the sample data follows a normal distribution by visualizing how the quantiles of the sample data match the quantiles of a theoretical normal distribution.
    """
    
    plt.figure(figsize=(10, 6))  # Create a figure with specified size
    
    # Plot histogram and KDE in the first subplot
    plt.subplot(1, 2, 1)
    sns.histplot(sample, kde=True)  # Use seaborn to plot histogram with KDE
    
    # Plot Q-Q plot in the second subplot
    plt.subplot(1, 2, 2)
    stat.probplot(sample, dist='norm', plot=pylab)  # Create a Q-Q plot comparing to normal distribution
    
    plt.show()  # Display the plots

```

A probability plot is a graphical technique used to assess whether a dataset follows a particular probability distribution. It can help in understanding how well a chosen distribution fits your data. Here are some common types of probability plots:

1. **Q-Q Plot (Quantile-Quantile Plot)**:
   - Compares the quantiles of your dataset to the quantiles of a theoretical distribution (e.g., normal distribution).
   - If the points lie approximately on a straight line, the data likely follows the theoretical distribution.

2. **P-P Plot (Probability-Probability Plot)**:
   - Compares the cumulative distribution function (CDF) of your dataset to the CDF of a theoretical distribution.
   - If the points lie close to the diagonal line, the data follows the theoretical distribution well.

3. **Probability Plot for Exponential Data**:
   - Used specifically for checking if the data follows an exponential distribution.
   - Similar to other plots but tailored to the exponential distribution's characteristics.

4. **Weibull Probability Plot**:
   - Used for checking if data follows a Weibull distribution, often used in reliability analysis.

To create a probability plot, you typically follow these steps:

1. **Choose the Distribution**: Decide which theoretical distribution you want to compare your data against.
2. **Compute Quantiles or CDF**: Calculate the quantiles or cumulative probabilities of your dataset and the theoretical distribution.
3. **Plot the Data**: Plot the quantiles or cumulative probabilities on a graph.
4. **Interpret the Plot**: Look at how closely the plotted points follow the reference line (e.g., the diagonal line). Deviations may suggest that the data does not fit the distribution well.
