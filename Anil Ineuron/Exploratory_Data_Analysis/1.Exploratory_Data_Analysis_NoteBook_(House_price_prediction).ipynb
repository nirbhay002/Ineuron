```python
#here we canoot find the detailed represenation of missing values 
features_with_nan =[features for features in dataset.columns if dataset[features].isnull().sum()>=1] #it basic gives columns
```

```python
features_with_nan #this are the features with nan values
```

```python
# i want to find the percentage of Nan values there 
dataset['Electrical'].isnull().mean()
```

```python
for feature in features_with_nan:
  print(feature, np.round(dataset[feature].isnull().mean(),4), '%missing Value')
```

```python
#let's plot some amazing diagram
data = dataset.copy()

for feature in features_with_nan:
  # let's make a vrariable that indicate 1 if the observation was missing 
  data[feature] = np.where(data[feature].isnull(),1,0)

  #lets calculate the mean salePrice where the information is missing or present
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House price Vs Year Sold')
```

```python
### here we will compare the differnce between all year features with salesprice
data = dataset.copy()
for feature in year_feature:
  if feature!= 'YrSold' :
     # because i have to subtract it from 
     data[feature] = data['YrSold']- data[feature] #Property age
     plt.scatter(data[feature],data['SalePrice'])
     plt.xlabel(feature)
     plt.ylabel('SalePrice')
     plt.show()
```

Observations

within all the yearwise information , it is clearly visible that the New houses are more costlier than the older one, lokks like pareto or power law Distribution

```python
### numerical variables - 2 Types
#1. continous variable or discrete variable

discrete_feature =[feature for feature in Numerical_features if len(dataset[feature].unique())<=25]
print(len(discrete_feature))
```

```python
data = dataset.copy()
## find the relationshipm between the discrete and the sales price 
for feature in discrete_feature:
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.show()
```

```python
## There is relation between the discrete feature and the saleprice
# continous variable
continous_feature = [feature for feature in Numerical_features if feature not in discrete_feature + year_feature]
```

```python
year_feature
```

```python
print(len(continous_feature))
```

```python
### Lets analyse the continous values by creating histograms to understand the distribution
data = dataset.copy()
for feature in continous_feature:
  data.groupby(feature)['SalePrice'].median().plot.hist()
  plt.xlabel(feature)
  plt.ylabel("SalePrice")
  plt.title(feature)
  plt.show()
  
```

```python
data = dataset.copy()
for feature in continous_feature:
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel("SalePrice")
  plt.title(feature)
  plt.show()
```

#EDA PART 2

```python
#we will be using logarthmic transformation
data = dataset.copy()
for feature in continous_feature:
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]= np.log(data[feature])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

```python
#Outliers
for feature in continous_feature:
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    data.boxplot(column = feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
#  the boxplot can help visualize the distribution of the continuous variable and
# the log transformation can help reduce the impact of extreme values in the distribution.
```
