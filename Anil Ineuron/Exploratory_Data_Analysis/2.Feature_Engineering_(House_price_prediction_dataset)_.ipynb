```python
# handling Missinmg value
features_nan =[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
for feature in features_nan:
  print("the feature is {} and missing values in {}%".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#the type of replcement i want to do , i can again create a seprate category called missing , and i can basically add it 
data = dataset.copy()
def replace_cat_features(dataset,categorical_features):
  data[categorical_features]=data[categorical_features].fillna('Missing')
  return data
#here we are adding a new constant ,through this new constant you will be getting a particular values
```

```python
replace_cat_features(data,categorical_features)
```

```python
data = dataset.copy()
def replace_cat_features(dataset,categorical_features):
  data[categorical_features]=data[categorical_features].fillna(data[categorical_features].mode())
  return data
```

```python
replace_cat_features(data,categorical_features)
```

```python
#Numerical values turn
#combining numerical features with nan values 
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtypes!='O' and 
                          dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
#here we are creating extra feature 
for feature in numerical_features_nan:
  # since there are outliers we are going to replace with median
  median_values = dataset[feature].median()
  ## create a new feature to capture nan value
  dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0) 
  # i also have one more feature that have importance of missing values , that will help us to crate the model more nicely
  dataset[feature].fillna(median_values,inplace=True)   #All nan values get filled by median values 
```
