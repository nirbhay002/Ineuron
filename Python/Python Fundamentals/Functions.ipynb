```python
def pattern(n):
  """This is a function which will help you to create lower bound pattern triangle"""
  for i in range(1,n+1):
    for j in range(1,i+1):
      print("*",end=" ")
    print("\n\r")


```

```python
pattern(5)
```

```python
def Create_Triangle(row_num):
  n= row_num
  k = n - 1
  for i in range(0, n):
    for j in range(0, k):
      print(end=" ")
    k = k - 1
    for j in range(0, i + 1 ):
      print("* ", end="")
    print("\n")
Create_Triangle(3)
```
